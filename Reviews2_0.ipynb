```
def counter_A(data,key_dict):
  data=str(data).lower()
  new_dict=defaultdict(int)
  for i in key_dict.keys():
    if " "+i+" " in data:
      new_dict[i] += 1
  return new_dict

def counter_num(df):
  reg=[]
  # for rows in df:
  y=re.findall('\d+\S*\s*',str(df))
  reg.append(y)
  return reg

# def counter_alphanum(data):
#   data=str(data).lower()
#   new_dict=defaultdict(int)
#   for i in alpha_list:
#     if i in data:
#       new_dict[i] += 1
#   return new_dict
```
